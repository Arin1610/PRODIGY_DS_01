### Data Visualization
Plotting Histogram of the data from 1960 - 2000


```python
years = df.columns[1:42]
```

```python
for i in years:
    fig = plt.figure(figsize = (5,5))
    plt.hist(df[i],color = 'cyan',bins = 10) #cyan color
    plt.xlabel(i)
    plt.show()
```

```python
#Sum the values for each year across all countries
total_values = df[years].sum()

#Plotting the bar chart
plt.figure(figsize = (20,30))
plt.barh(years,total_values,color = 'red')
plt.xlabel('Total Values')
plt.ylabel('Year', size = 20)
plt.title('Total values per Year', size = 20)
plt.show()
```

### Plotting Indicators of top 10 Countries for different years with Base year 



```python
country_1960 = df.sort_values(by = '1960').head(10)
country_1960
```

```python
country_1960_t = country_1960.set_index('Country Name').T
for year,data_values in country_1960_t.iterrows():
    fig = plt.figure(figsize = (10,5))
    sns.barplot(x = data_values.index , y = data_values.values)
    plt.xlabel('Countries')
    plt.ylabel('Data Values')
    plt.title(f"{year} - Data Values From 1960 to 2000")
    plt.xticks(rotation = 90)
    plt.show()
```
